```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the data
data_url = 'marketing_customer_analysis.csv'
df = pd.read_csv(data_url)

# Show DataFrame info
df.info()
```

```python
# Describe DataFrame
df.describe(include='all')
```

```python
# Show a plot of the total number of responses
plt.figure(figsize=(8,6))
sns.countplot(x='Response', data=df)
plt.title('Total Number of Responses')
plt.show()
```

```python
# Show a plot of the response rate by the sales channel
plt.figure(figsize=(10,6))
sns.countplot(x='Sales Channel', hue='Response', data=df)
plt.title('Response Rate by Sales Channel')
plt.show()
```

```python
# Show a plot of the response rate by the total claim amount
plt.figure(figsize=(10,6))
sns.boxplot(x='Response', y='Total Claim Amount', data=df)
plt.title('Response Rate by Total Claim Amount')
plt.show()
```

```python
# Show a plot of the response rate by income
plt.figure(figsize=(10,6))
sns.boxplot(x='Response', y='Income', data=df)
plt.title('Response Rate by Income')
plt.show()
```

```python
# Get the numeric data into dataframe called numerical
numerical = df.select_dtypes(include=np.number)

# Get categorical columns in a dataframe called categoricals
categoricals = df.select_dtypes(include=object)

numerical.head(), categoricals.head()
```

```python
# Use seaborn library to construct distribution plots for the numerical variables
for column in numerical.columns:
    sns.displot(numerical[column])
    plt.show()
```

```python
# Use Matplotlib to construct histograms
for column in numerical.columns:
    plt.hist(numerical[column], bins=30)
    plt.title(column)
    plt.show()
```

```python
# Compute the skewness for each, and add a comment with your findings.
skewness = numerical.apply(lambda x: x.skew()).sort_values(ascending=False)
skewness
```

```python
# Check the multicollinearity between the features
correlations = numerical.corr()
correlations
```

```python
# If there is a pair of columns that show a high correlation between them (greater than 0.9),
# drop the one that is less correlated with the column total_claim_amount.
# Write code for both the correlation matrix.
# If there is no pair of features that have a high correlation, then do not drop any features.

# Create correlation matrix
corr_matrix = numerical.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find index of feature columns with correlation greater than 0.9
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]

# Drop features
numerical = numerical.drop(numerical[to_drop], axis=1)

numerical.head()
```

```python
# Plot the heatmap of the correlation matrix after the filtering
plt.figure(figsize=(10,8))
sns.heatmap(numerical.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
